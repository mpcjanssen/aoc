```tcl
tcl::tm::path add [file join . modules]
package require util
```

```tcl
set data [read-input day10]
```

```tcl
set ex1 {......#.#.
#..#.#....
..#######.
.#.#.###..
.#..#.....
..#....#.#
#..#....#.
.##.#..###
##...#..#.
.#....####}
```

```tcl
set grid [lmap x [split $ex1 \n]  {split $x {}}]
```

```tcl
set points {} ; foreach x [range 0 9] {foreach  y [range 0 9] {if {[lindex $grid $x $y] eq "#"} {lappend points [list $x $y]}}}
```

```tcl
set points
```

```tcl
namespace import tcl::mathfunc::atan2
```

```tcl
atan2 0 0
```

```tcl
set angles {}
foreach p1 $points {
    lassign $p1 x1 y1
    foreach p2 $points {

        if {$p1 eq $p2} continue
        lassign $p2 x2 y2
        puts "$x1:$y1 - $x2:$y2"
        set dx [expr {$x2 - $x1}]
        set dy [expr {$y2 - $y1}]
        puts [dict lappend angels $p1 [atan2 $dx $dy]]
    }
}
```

```tcl
puts $angles
```
